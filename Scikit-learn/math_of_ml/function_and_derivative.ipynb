```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
%matplotlib inline
```

```python
# f(x) = x
def linear(x):
    return x

lin_x, lin_y = [], []

for i in range(-10, 11, 1):
    lin_x.append(i)
    lin_y.append(linear(i))
    
plt.figure(figsize=(12, 5))
plt.plot(lin_x, lin_y)
plt.grid(True)
plt.show();
```

```python
# f(x) = 1 / (1 + np.exp ** -x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

sig_x, sig_y = [], []

for i in range(-10, 11, 1):
    sig_x.append(i)
    sig_y.append(sigmoid(i))
    
plt.figure(figsize=(12, 5))
plt.plot(sig_x, sig_y)
plt.grid(True)
plt.show();
```

```python
derlin_x, derlin_y = [], []

for i in range(-10, 11, 1):
    derlin_x.append(i)
    derlin_y.append(derivative(linear, i))

plt.figure(figsize=(12, 5))
plt.plot(derlin_x, derlin_y)
plt.grid(True)
plt.show();
```

```python
dersig_x, dersig_y = [], []

for i in range(-10, 11, 1):
    dersig_x.append(i)
    dersig_y.append(derivative(sigmoid, i, dx=1e-6))
    
plt.figure(figsize=(12, 5))
plt.plot(dersig_x, dersig_y)
plt.grid(True)
plt.show();
```
