```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from numpy.random import exponential
```

```python
fig, axes = plt.subplots()
axes.plot(exponential(5, 20))
axes.set_title('Graph')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))

for row, row_ax in enumerate(axes):
    for col, ax in enumerate(row_ax):
        ax.plot(exponential(col, 20))
        ax.set_title(f'Canvas col {col + 1} row {row + 1}')
fig.tight_layout()
```

```python
fig.savefig('all_results.png')
```
