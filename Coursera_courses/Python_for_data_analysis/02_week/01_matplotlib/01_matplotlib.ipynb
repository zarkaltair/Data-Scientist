```python
import pandas as pd
from numpy.random import exponential
```

```python
df = pd.DataFrame({'x': range(20), 'y': exponential(10, 20)})
```

```python
%matplotlib inline
```

```python
df.y.hist();
```

```python
import matplotlib.pyplot as plt
```

```python
data1 = exponential(5, 20)
data2 = exponential(6, 20)
```

```python
plt.plot(data1, label='Player One')
plt.scatter(range(len(data2)), data2, color='red', label='Player Two')
plt.title('Results', fontdict={'fontsize': 20})
plt.xlabel('Number try')
plt.ylabel('Result')
plt.xticks(range(0, 20, 4))
plt.legend();
```

```python
plt.savefig('results.pdf')
```
