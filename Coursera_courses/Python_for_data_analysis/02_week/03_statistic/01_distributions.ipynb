## Discrete distributions

```python
rv_bernoulli = stats.bernoulli(p=0.3)
```

```python
rv_bernoulli.rvs(14)
```

```python
rv_binom = stats.binom(100, p=0.9)
```

```python
rv_binom.rvs(8)
```

```python
plt.hist(rv_binom.rvs(80), bins=10);
```

## Continuous distributions

```python
a, b = 5, 10

rv_uniform = stats.uniform(a, b-a)
```

```python
rv_uniform.cdf(5.5)
```

```python
rv_uniform.cdf(11)
```

```python
rv_uniform.pdf(7)
```

```python
x = np.linspace(a - 2, b + 2, 100)
cdf = rv_uniform.cdf(x)
plt.plot(x, cdf)

plt.ylabel('F(x)')
plt.xlabel('X')
plt.ylim([0, 1.5])
plt.title('Cumulative distribution function for uniform');
```

```python
x = np.linspace(a - 2, b + 2, 1000)
pdf = rv_uniform.pdf(x)
plt.plot(x, pdf)

plt.xlabel('f(x)')
plt.ylabel('x')
plt.xlim([4, 12])
# plt.ylim([0, 1])
plt.title('PDF for uniform');
```

## Normal distributions

```python
mu, sigma = 2, 0.5

rv_norm = stats.norm(loc=mu, scale=sigma)
```

```python
rv_norm.rvs(17)
```

```python
x = np.linspace(0, 4, 100)
cdf = rv_norm.cdf(x)
plt.plot(x, cdf)

plt.ylabel('F(x)')
plt.xlabel('X');
```

```python
x = np.linspace(0, 4, 100)
pdf = rv_norm.pdf(x)
plt.plot(x, pdf)

plt.xlabel('X')
plt.ylabel('F(x)');
```

```python
means = [1, 5, 10]
mu = 10
sigmas = [1, 2, 4, 6]
for sigma in sigmas:
    rv_norm = stats.norm(loc=mu, scale=sigma)
    x = np.linspace(0, 20, 200)
    pdf = rv_norm.pdf(x)
    plt.plot(x, pdf, label=sigma)
plt.legend();
```
