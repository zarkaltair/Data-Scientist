```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 8)
```

### Экспоненциальное распределение

```python
data = sts.expon(0)
data_rvs = data.rvs(1000)
```

### Строим гистограмму по выборке и теоретическую плотность

```python
plt.hist(data_rvs, bins=100, normed=True, histtype='stepfilled', alpha=0.5, label='Hist')

x = np.linspace(0, 10, 1000)
plt.plot(x, data.pdf(x), 'r', lw=2, label='Theor pdf')
plt.legend(loc='best', frameon=True)
plt.xlabel('x')
plt.ylabel('f(x)');
```

### Генерируем распределение из выборочных средник при n=5

```python
arr_5 = []
for i in range(0, 1000):
    arr_5.append(data.rvs(5).mean())
```

```python
plt.hist(arr_5, bins=100, density=True, histtype='stepfilled', alpha=0.5, label='Hist_n_5')

plt.legend(loc='best', frameon=True)
plt.xlabel('x')
plt.ylabel('f(x)')
sns.kdeplot(arr_5);
```

### Вывод

В результате исследования распределения средних значений 50 экспоненциальных случайных величин мы показали, что и среднее значение, и дисперсия данного распределения достаточно близки к теоретическим значениям.
